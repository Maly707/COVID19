```python
#Dependancy 
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import import_ipynb
import Covid19_test
TotalCovid19Cases = Covid19_test.TotalCovid19Cases
```

```python
TotalCovid19Cases.plot(
    title='Chinese Quarantine highlight', 
    lw=2,
    colormap='jet', 
    marker='.', 
    markersize=12, 
    figsize=(20,5), y=['China_confirmid']).legend(["Confirmid"])
plt.axvspan('1/23/20', TotalCovid19Cases.tail(1).index.item(), facecolor='g', alpha=0.5)
plt.show()

TotalCovid19Cases.plot(
    title='Italy lockdown highlight', 
    lw=2,
    colormap='jet', 
    marker='.', 
    markersize=12, 
    figsize=(20,5), y=['Italy_confirmid']).legend(["Confirmid"])
plt.axvspan('3/9/20', TotalCovid19Cases.tail(1).index.item(), facecolor='y', alpha=0.5)
plt.show()

TotalCovid19Cases.plot(
    title='USA emergency state highlight', 
    lw=2,
    colormap='jet', 
    marker='.', 
    markersize=12, 
    figsize=(20,5), y=['US_confirmid']).legend(["Confirmid"])
plt.axvspan('3/13/20', TotalCovid19Cases.tail(1).index.item(), facecolor='r', alpha=0.5)
plt.show()
```
