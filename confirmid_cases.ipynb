```python
#Dependancy 
import pandas as pd
```

```python
covid19ConfirmedURL="https://data.humdata.org/hxlproxy/api/data-preview.csv?url=https%3A%2F%2Fraw.githubusercontent.com%2FCSSEGISandData%2FCOVID-19%2Fmaster%2Fcsse_covid_19_data%2Fcsse_covid_19_time_series%2Ftime_series_covid19_confirmed_global.csv&filename=time_series_covid19_confirmed_global.csv"
ConfirmedCases = pd.read_csv(covid19ConfirmedURL)
```

```python
totalCasesPerDay = ConfirmedCases[ConfirmedCases['Country/Region'].isin(['China', 'Italy','US'])].drop(['Province/State', 'Lat', 'Long'], axis=1)
totalCasesPerDay = totalCasesPerDay.groupby('Country/Region').sum()
totalCasesPerDay = totalCasesPerDay.T
totalCasesPerDay.tail()
```

```python
totalCasesPerDay.describe()
```

```python
totalCasesPerDay.plot(lw=2, colormap='jet', marker='.', markersize=12, figsize=(20,5))
```

```python
newCasesPerDay = totalCasesPerDay.diff()
newCasesPerDay.tail()
```

```python
newCasesPerDay.describe()
```

```python
newCasesPerDay.plot(lw=2, colormap='jet', marker='.', markersize=12, figsize=(20,5))
```
