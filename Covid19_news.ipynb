```python
import requests
from bs4 import BeautifulSoup
from csv import writer
from datetime import datetime
from dateutil.parser import parse
import numpy as np
import pandas as pd
import csv
import re
import os
```

```python
def matchKeyword(s, keywords):
    words_re = re.compile("|".join(keywords).lower())
    if words_re.search(s.lower()):
        return True
    return False

def applyFunc(s):
    china = ['China', 'Chinese']
    if matchKeyword(s, china):
        return china[0]
    
    usa = [' USA ', 'united state', 'U.S.', ' USA.']
    if matchKeyword(s, usa):
        return usa[0]
    
    italy = ['Italy']
    if matchKeyword(s, italy):
        return usa[0]
    
    return ''
```

```python
response = requests.get('https://www.businessinsider.com/coronavirus-pandemic-timeline-history-major-events-2020-3')
soup = BeautifulSoup(response.text, 'html.parser')
# soup.find_all("div", class_="stylelistrow")
posts = soup.find_all('h2', class_="slide-title-text")
result = np.array([])
```

```python
filename = 'data/topStory.csv'
if os.path.exists(filename):
  os.remove(filename)
with open(filename, 'a+', newline='', encoding="utf8") as write_obj:
        csv_writer = writer(write_obj)
        csv_writer.writerow(['Date','TopStory','country'])
        for post in posts:
            country = ''
            story = post.text.split(":")
            dtime = story[0]
            if isinstance(dtime.split("-"), list):
                dtime = dtime.split(":")[0]
            try:
                date = parse(dtime).strftime('%m/%d/%y')
            except ValueError:
                continue 
            topStory = story[1]
            country = applyFunc(topStory)
            csv_writer.writerow(np.append(result, [date,topStory,country], axis=0))
```
